```python
def mean(arr):
    s = 0
    for i in arr:
        s += int(i)
    return (s/len(arr))

def median(arr):
    arr.sort()
    arr_len=len(arr)
    if (arr_len % 2 == 0):
        mid = arr_len//2
        median = (arr[mid-1] + arr[mid])/2.0
        return median
    else:
        mid = arr_len//2
        median = arr[mid]
        return median
    
def mode(arr):
    counts = {}
    for num in arr:
        counts[num] = counts.get(num, 0) + 1

    max_freq = max(counts.values())

    modes = []
    for num, count in counts.items():
        if count == max_freq:
            modes.append(num)
    return modes

if __name__ == "__main__":
    arr = []
    arr_len=int(input("Enter number of Elements:"))
    print(f"Enter {arr_len} elements:")
    for i in range(arr_len):
        no = input("")
        arr.append(float(no))

    print("Mean:",mean(arr))
    print("Median:",median(arr))
    print("Mode(s):",mode(arr))
```
