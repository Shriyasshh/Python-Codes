```python
def nearly_equal(s1,s2):
    if (len(s1) != len(s2)):
        return False
    else:
        mutate = 0
        for i in range(len(s1)):  
            if (s1[i] != s2[i]):
                mutate += 1
            if mutate > 1:
                return False
        return True
    
    
    

s1=(input("Enter a word:"))
s2= (input("Enter 2nd word:"))
print(nearly_equal(s1,s2))
```
