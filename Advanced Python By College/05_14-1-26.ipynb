```python
from datetime import datetime, time
```

```python
date_str = input("Enter a date (DD-MM-YYYY) or (DD/MM/YYYY): ")
```

```python


if "-" in date_str:
    dt = datetime.strptime(date_str, "%d-%m-%Y").date()
    print(dt.strftime("%d/%m/%Y"))
    print(dt.strftime("%d %B %Y"))
elif "/" in date_str: 
    dt = datetime.strptime(date_str, "%d/%m/%Y").date()
    print(dt.strftime("%d-%m-%Y"))
    print(dt.strftime("%d %B %Y"))
else:
    print("Invalid")
```

```python
t = datetime.now().time()
format1 =  t.strftime("%H:%M:%S")
format2 =  t.strftime("%I:%M:%S")
print(format1)
print(format2)
```
