```python
class Laptop():
    def putdata(self):
        self.brand=input("Enter Laptop Brand:")
        self.price=int(input("Enter Price:"))
        self.processor=input("Enter name of processor in Laptop:")
    
    def display(self):
        print("Brand:",self.brand)
        print("Price:",self.price)
        print("Processor:",self.processor)

l1= Laptop()
l1.putdata()
l1.display()

l2=Laptop()
l2.putdata()
l2.display()
```

```python
class Hospital():
    def PatientDetail(self):
        self.name=input("Enter Patient name:")
        self.age=int(input("Enter Patient age:"))
        self.disease=input("Enter Patient's Disease:")
    
    def display(self):
        print("Patient Name:",self.name)
        print("Age:",self.age)
        print("Disease:",self.disease)

p1 = Hospital()
p1.PatientDetail()
p1.display()

p2=Hospital()
p2.PatientDetail()
p2.display()

p3=Hospital()
p3.PatientDetail()
p3.display()
```
