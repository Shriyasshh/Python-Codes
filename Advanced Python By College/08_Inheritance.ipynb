# Single Inheritance

```python
class parent:
    def fun1(self):
        print("This is parent class")
    
class child(parent):
    def fun2(self):
        print("This is child class")


obj = child()
obj.fun2()
obj.fun1()
```

```python
class Employee:
    def __init__(self,name,salary):
        self.name= name
        self.salary= salary

class Dept(Employee):
    def __init__(self, name, salary,dept):
        super().__init__(name, salary)
        self.dept= dept
    def print(self):
        print(f"Name:{self.name} Salary:{self.salary} Dept:{self.dept}")

E=Dept("john",30000,"IT")
E.print()
```

# Multiple Inheritance

```python
class parent1:
    def fun1(self):
        print("This is parent1 class")
    
class parent2:
    def fun3(self):
        print("This is parent2 class")

class child(parent1,parent2):
    def fun2(self):
        print("This is child class")

obj = child()
obj.fun2()
obj.fun3()
obj.fun1()
```

```python
class Rectangle:
    def area_rectangle(self,l,b):
        return l*b
class Circle:
    def area_circle(self,r):
        return 3.14*r*r
class Area(Rectangle,Circle):
    def display(self):
        print("Rectangle Area:",self.area_rectangle(5,4))
        print("Circle Area:",self.area_circle(3))
obj=Area()
obj.display()
```

# Multilevel Inheritance

```python
class parent1:
    def fun1(self):
        print("This is parent1 class")
    
class parent2(parent1):
    def fun3(self):
        print("This is parent2 class")

class child(parent2):
    def fun2(self):
        print("This is child class")

obj = child()
obj.fun2()
obj.fun3()
obj.fun1()
```

```python
class Shape:
    def area_rectangle(self,l,b):
        return l*b
    
class Rectangle(Shape):
    def perimeter(self,l,b):
        return 2*(l+b)
    
class Box(Rectangle):
    def volume(self,l,b,h):
        return self.area_rectangle(l,b)*h
obj=Box()
print("Area:",obj.area_rectangle(4,5))
print("Perimeter:",obj.perimeter(4,5))
print("Volume:",obj.volume(4,5,6))
```

# Herirarchy Inheritance

```python
class Parent:
    def fun1(self):
        print("This is parent class")

class Child1(Parent):
    def fun2(self):
        print("This is child 1 class")

class Child2(Parent):
    def fun3(self):
        print("This is child 2 class")

obj1 = Child1()
obj2 = Child2()
obj1.fun1()
obj1.fun2()
obj2.fun1()
obj2.fun3()
```

```python
class Number:
    def get_number(self,n):
        return n
class Square(Number):
    def square(self,n):
        return self.get_number(n)**2
class Cube(Number):
    def cube(self,n):
        return self.get_number(n)**3
class Factorial(Number):
    def factorial(self,n):
        f=1
        for i in range(1,n+1):
            f*=i
        return f
obj1=Square()
obj2=Cube()
obj3=Factorial()
print("Square:",obj1.square(4))
print("Cube:",obj2.cube(3))
print("Factorial:",obj3.factorial(5))
```
