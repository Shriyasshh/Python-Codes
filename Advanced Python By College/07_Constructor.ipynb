```python
class ATM():
    def __init__(self):
        self.AccHold_Name = input("Enter Name of Account Holder:")
        self.Acc_No = int(input("Enter Account Number:"))
        self.Balance = float(input("Enter Balance:"))
    
    def deposit(self):
        amount = float(input("Enter Amount to be deposited:"))
        self.Balance+=amount
        print("Balance after deposit:",self.Balance)

    def check_balance(self):
        print("Balance:",self.Balance)
    
    def withdraw(self):
        amount =float (input("Enter Amount to be withdrawn:"))
        if self.Balance >= amount:
            self.Balance-=amount
            print("Balance after withdrawal:",self.Balance)
        else:  
            print("Insufficient Balance")
    
    def delete_account(self):
        self.AccHold_Name = None
        self.Acc_No = None
        self.Balance = None
        print("Account Deleted")


        
person1 = ATM()
person1.deposit()
person1.check_balance()
person1.withdraw()
person1.delete_account()
person1.check_balance()
```

```python
class Calculator():
    def __init__(self, num1, num2, oper):
        self.num1 = num1
        self.num2 = num2
        self.oper = oper
    
    def calculate(self):
        if self.oper == '+':
            return self.addition()
        elif self.oper == '-':
            return self.sub()
        elif self.oper == '*':
            return self.product()
        elif self.oper == '/':
            return self.division()
        else:
            return 'Invalid Operator'

    def addition(self):
        return self.num1 + self.num2
    
    def sub(self):
        return self.num1 - self.num2
    
    def product (self):
        return self.num1 * self.num2
    
    def division(self):
        if self.num2 !=0:
            return self.num1 / self.num2
        else:
            return 'Cannot Divide by Zero'
        
cal = Calculator(10,0,'/')
print(cal.calculate())
```

```python
class Employee():
   def  __init__(self, name,salary):
      self.name = name
      self.salary = salary


class Manager(Employee):
   
   def __init__(self, name, salary,dept):
      Employee.__init__(self,name, salary)
      self.dept = dept
   def info(self):
      print(f"Manager name: {self.name}, Salary: {self.salary}, Department: {self.dept}")

class Executive(Manager):
   def __init__(self, name, salary, dept):
      super().__init__(name, salary, dept)
   
   def info(self):
      print("Executive")
      super().info()

E1 = Employee("Bob",6000)
obj1 = Manager("John",50000, 'IT')
o= Executive("Rob",500,"Advertising")
o.info()
print(E1.name)
print(E1.salary)
print(obj1.name)
print(obj1.salary)
print(obj1.dept)
```
